## Age
Values 17 through 90, skewed to the right since there are fewer seniors than any other age group. Looking at the value counts, we can see the number of instances trickle out as we get into the high 80s.

## Education-num
The education-num represents a bunch of different education brackets, High-school Grad starting at 9. As you can see 75% or more of people have at least a high school education and the mean is right above an Associate's degree. Education is left-skewed so the outliers are towards lower education

## Capital Gain


```python
qual_vars = ['Name', 'Zip', 'Workclass', 'Occupation', 'Race', 'Sex', 'Education', 'Country', 'Target']
qual_dist = [data.value_counts(subset=[qual_var], normalize=True) for qual_var in qual_vars]
#The normalized value_counts give us a general idea of the distribution of the qualitative variables


#I set this so that the print out is cleaner
pd.options.display.max_rows = 16
```

```python
printout = [str(qual_vars) for qual_vars in qual_dist]
for i in range(len(printout)):
    printout[i] += "\n"*max(0,18-len(printout[i].split("\n")))
    #This allows us to numpy.stack and create a uniform printout
    #print(len(str(printout[i]).split('\n')))
    
#These sections are hand-picked to make all the values fit in sections
stacked_section1 = np.stack([line.split('\n') for line in printout[:3]], axis=-1)
section1 = np.array(["\t".join(line) for line in stacked_section1])
stacked_section2 = np.stack([line.split('\n') for line in printout[3:-3]], axis=-1)
section2 = np.array(["\t".join(line) for line in stacked_section2])
stacked_section3 = np.stack([line.split('\n') for line in printout[-3:]], axis=-1)
section3 = np.array(["\t".join(line) for line in stacked_section3])

#Have to hand pick these values too, but it all works out
section1[0] = "Name\t\t\t\tZip\t\t\tWorkclass"
section2[0] = "Occupation\t\t\tRace\t\t\t\tSex"
section3[0] = "Education\t\t\tCountry\t\t\t\t\tTarget"

for i in section1: print(i)
for i in section2: print(i)
for i in section3: print(i)
```
